```python
import requests

response = requests.post('https://fbrapi.com/generate_api_key')
api_key = response.json()['api_key']
print("API Key:", api_key)
```

```python
import time 

# First get countries
url = "https://fbrapi.com/countries"
headers = {"X-API-Key": api_key}

response = requests.get(url, headers=headers)
countries = response.json()
print("Countries:", countries)
time.sleep(6)

# Define the countries we're looking for
target_countries = ['England', 'Spain', 'Italy', 'France', 'Germany']
top_5_country_codes = []

# Extract country codes from the data
for country in countries['data']:
    if country['country'] in target_countries:
        top_5_country_codes.append(country['country_code'])

print("Top 5 country codes:", top_5_country_codes)
```

```python
# Get league ids for each country
target_leagues = ['Premier League', 'Ligue 1', 'Bundesliga', 'Serie A', 'La Liga']
top_5_league_ids = []

for country_code in top_5_country_codes:
    url = "https://fbrapi.com/leagues"
    params = {"country_code": country_code}  # Changed from country_code to country_id
    response = requests.get(url, headers=headers, params=params)
    leagues = response.json()
    
    # Navigate through the nested structure
    if 'data' in leagues:
        for league_type in leagues['data']:
            if 'leagues' in league_type:
                for league in league_type['leagues']:
                    if league['competition_name'] in target_leagues:
                        top_5_league_ids.append(league['league_id'])
    
    time.sleep(6)

print("Top 5 league IDs:", top_5_league_ids)
```
